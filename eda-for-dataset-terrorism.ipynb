```python
#now we will visualize missing data in dataset through heatmap 
missing_data = DataFrame.columns[:30]
colors=['#0000FF','#FF69B4']
sns.heatmap(DataFrame[missing_data].isnull(),cmap=sns.color_palette(colors))
```

the above heatmap will show the missing values in each group where pink color is indicating missing values and blue color otherwise.

now lets find out missing value through percentage data list 

```python
#missing data percentage list
for col in DataFrame.columns:
    missing =np.mean(DataFrame[col].isnull())
    print('{} - {}%'.format(col,round(missing*100)))
```

first we will fill null values of non numeric data 

```python
DataFrame["approxdate"]=DataFrame["approxdate"].fillna('None')
DataFrame["resolution"]=DataFrame["resolution"].fillna('None')
DataFrame["country_txt"]=DataFrame["country_txt"].fillna('None')
DataFrame["region_txt"]=DataFrame["region_txt"].fillna('None')
DataFrame["provstate"]=DataFrame["provstate"].fillna('None')
DataFrame["city"]=DataFrame["city"].fillna('None')
DataFrame["location"]=DataFrame["location"].fillna('None')
DataFrame["summary"]=DataFrame["summary"].fillna('None')
DataFrame["alternative_txt"]=DataFrame["alternative_txt"].fillna('None')
DataFrame["attacktype1_txt"]=DataFrame["attacktype1_txt"].fillna('None')
DataFrame["gname"]=DataFrame["gname"].fillna('None')
DataFrame["gsubname"]=DataFrame["gsubname"].fillna('None')
DataFrame["motive"]=DataFrame["motive"].fillna('None')
DataFrame["claimmode_txt"]=DataFrame["claimmode_txt"].fillna('None')
DataFrame["weapdetail"]=DataFrame["weapdetail"].fillna('None')
DataFrame["propextent_txt"]=DataFrame["propextent_txt"].fillna('None')
DataFrame["propcomment"]=DataFrame["propcomment"].fillna('None')
DataFrame["divert"]=DataFrame["divert"].fillna('None')
DataFrame["addnotes"]=DataFrame["addnotes"].fillna('None')
DataFrame["scite1"]=DataFrame["scite1"].fillna('None')
DataFrame["scite2"]=DataFrame["scite2"].fillna('None')
DataFrame["scite3"]=DataFrame["scite3"].fillna('None')
DataFrame["dbsource"]=DataFrame["dbsource"].fillna('None')
DataFrame["related"]=DataFrame["related"].fillna('None')
```

Now we will assign median value to null values of numeric data 

```python
DataFrame["eventid"]=DataFrame["eventid"].fillna(DataFrame["eventid"].median())
DataFrame["iyear"]=DataFrame["iyear"].fillna(DataFrame["iyear"].median())
DataFrame["imonth"]=DataFrame["imonth"].fillna(DataFrame["imonth"].median())
DataFrame["iday"]=DataFrame["iday"].fillna(DataFrame["iday"].median())
DataFrame["extended"]=DataFrame["extended"].fillna(DataFrame["extended"].median())
DataFrame["country"]=DataFrame["country"].fillna(DataFrame["country"].median())
DataFrame["region"]=DataFrame["region"].fillna(DataFrame["region"].median())
DataFrame["latitude"]=DataFrame["latitude"].fillna(DataFrame["latitude"].median())
DataFrame["longitude"]=DataFrame["longitude"].fillna(DataFrame["longitude"].median())
DataFrame["specificity"]=DataFrame["specificity"].fillna(DataFrame["specificity"].median())
DataFrame["vicinity"]=DataFrame["vicinity"].fillna(DataFrame["vicinity"].median())
DataFrame["crit1"]=DataFrame["crit1"].fillna(DataFrame["crit1"].median())
DataFrame["crit2"]=DataFrame["crit2"].fillna(DataFrame["crit2"].median())
DataFrame["crit3"]=DataFrame["crit3"].fillna(DataFrame["crit3"].median())
DataFrame["guncertain1"]=DataFrame["guncertain1"].fillna(DataFrame["guncertain1"].median())
DataFrame["guncertain2"]=DataFrame["guncertain2"].fillna(DataFrame["guncertain2"].median())
DataFrame["guncertain3"]=DataFrame["guncertain3"].fillna(DataFrame["guncertain3"].median())
DataFrame["individual"]=DataFrame["individual"].fillna(DataFrame["individual"].median())
DataFrame["nperps"]=DataFrame["nperps"].fillna(DataFrame["nperps"].median())
DataFrame["claimed"]=DataFrame["claimed"].fillna(DataFrame["claimed"].median())
DataFrame["compclaim"]=DataFrame["compclaim"].fillna(DataFrame["compclaim"].median())
```
